```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
print(plt.style.available)
```

## Matlab

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x+1,'r--')
ax.plot(x, x+2,'g-')
ax.plot(x, x+3,'y:')
ax.plot(x, x+4,'bo')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x+1,color= 'blue', alpha= 0.2)
ax.plot(x, x+2,color = 'red', linewidth = 10)
ax.plot(x, x+3,color = 'green',linestyle = '--')
ax.plot(x, x+4,color = 'yellow', marker = 'X')
```
